```python
#Function to train a classifier, test the classifier on other data sets, output accuracy score results 
#It outputs the accuracy scores at each temperature in a list form 
#It also prints all the accuracy scores and produces a loss curve
def traintestclassifier(max_iterations, n_iterations_no_change, nodes_per_layer, trainingData, trainingLabels):
    clf = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes=(nodes_per_layer,nodes_per_layer,nodes_per_layer,nodes_per_layer,nodes_per_layer), max_iter=max_iterations, tol=0, n_iter_no_change=n_iterations_no_change)

    #training network on base data set 
    clf.fit(trainingData, trainingLabels)

    #Printing graph of loss curve
    plt.plot(clf.loss_curve_)

    #using the T0.45 trained neural network to predict the labels for the other data sets 
    test_predictT045 = clf.predict(test_dataT045) 
    test_predictT047 = clf.predict(test_dataT047) 
    test_predictT051 = clf.predict(test_dataT051) 
    test_predictT053 = clf.predict(test_dataT053) 
    test_predictT058 = clf.predict(test_dataT058) 
    test_predictT07 = clf.predict(test_dataT07)
    test_predictT073 = clf.predict(test_dataT073)
    test_predictT076 = clf.predict(test_dataT076)
    test_predictT078 = clf.predict(test_dataT078)
    test_predictT081 = clf.predict(test_dataT081)
    test_predictT083 = clf.predict(test_dataT083)   
    test_predictT086 = clf.predict(test_dataT086)    
    test_predictT10 = clf.predict(test_dataT10) 
    test_predictT12 = clf.predict(test_dataT12) 
    test_predictT14 = clf.predict(test_dataT14) 
    test_predictT18 = clf.predict(test_dataT18)
    test_predictT20 = clf.predict(test_dataT20)

    #calculating the accuracy scores 
    scoreT045 = round((accuracy_score(test_predictT045, test_labelsT045)), 4)
    scoreT047 = round((accuracy_score(test_predictT047, test_labelsT047)),4)
    scoreT051 = round((accuracy_score(test_predictT051, test_labelsT051)),4)
    scoreT053 = round((accuracy_score(test_predictT053, test_labelsT053)),4)
    scoreT058 = round((accuracy_score(test_predictT058, test_labelsT058)),4)
    scoreT07 = round((accuracy_score(test_predictT07, test_labelsT07)),4)
    scoreT073 = round((accuracy_score(test_predictT073, test_labelsT073)),4)
    scoreT076 = round((accuracy_score(test_predictT076, test_labelsT076)),4)
    scoreT078 = round((accuracy_score(test_predictT078, test_labelsT078)),4)
    scoreT081 = round((accuracy_score(test_predictT081, test_labelsT081)),4)
    scoreT083 = round((accuracy_score(test_predictT083, test_labelsT083)),4)
    scoreT086 = round((accuracy_score(test_predictT086, test_labelsT086)),4)
    scoreT10 = round((accuracy_score(test_predictT10, test_labelsT10)),4)
    scoreT12 = round((accuracy_score(test_predictT12, test_labelsT12)),4)
    scoreT14 = round((accuracy_score(test_predictT14, test_labelsT14)),4)
    scoreT18 = round((accuracy_score(test_predictT18, test_labelsT18)),4)
    scoreT20 = round((accuracy_score(test_predictT20, test_labelsT20)),4)

    accuracies = [scoreT045, scoreT047, scoreT051, scoreT053, scoreT058, scoreT07, scoreT073, scoreT076, scoreT078, scoreT081, scoreT083, scoreT086, scoreT10, scoreT12, scoreT14, scoreT18, scoreT20]


    #Printing accuracy scores for each data set
    print("Accuracy Scores: \n")
    print("T_0.45: " + str(scoreT045))
    print("T_0.47: " + str(scoreT047))
    print("T_0.51: " + str(scoreT051))
    print("T_0.53: " + str(scoreT053))
    print("T_0.58: " + str(scoreT058))
    print("T_0.7: " + str(scoreT07))
    print("T_0.73: " + str(scoreT073))
    print("T_0.76: " + str(scoreT076))
    print("T_0.78: " + str(scoreT078))
    print("T_0.81: " + str(scoreT081))
    print("T_0.83: " + str(scoreT083))
    print("T_0.86: " + str(scoreT086))
    print("T_1.0: " + str(scoreT10))
    print("T_1.2: " + str(scoreT12))
    print("T_1.4: " + str(scoreT14))
    print("T_1.8: " + str(scoreT18))
    print("T_2.0: " + str(scoreT20))

    return accuracies
```
